#### 02. Decision Tree Implementation on Iris Dataset

```python
x_train
```

```python
df = pd.DataFrame(x_train)
```

```python
df.columns = ['sep_len','sep_wid','pet_len','pet_wid']
```

```python
df
```

```python
# converting continuous values to discreet values
# For simplicity, assuming just two values discreet values for each feature 
# Low --> Less than Mean of Data
# High --> Greater than Mean of Data
# Same can be used for test data set as well

def discretize(value,*mean_val):

    # if any value is less than mean value, call it Low
    # else call it High

    if value <= mean_val[0]:
        return 'Low'
    else:
        return 'High'
        
sep_len_mean = df['sep_len'].mean()
df['sep_len'] = df['sep_len'].apply(discretize, args= [sep_len_mean])
sep_wid_mean = df['sep_wid'].mean()
df['sep_wid'] = df['sep_wid'].apply(discretize, args= [sep_wid_mean])
pet_len_mean = df['pet_len'].mean()
df['pet_len'] = df['pet_len'].apply(discretize, args= [pet_len_mean])
pet_wid_mean = df['pet_wid'].mean()
df['pet_wid'] = df['pet_wid'].apply(discretize, args= [pet_wid_mean])
```

```python
arr = df.to_numpy()    # our input array to the build tree function
```

```python
def entropy(labels):             # gives entropy prior to splitting

    unique, counts = np.unique(labels, return_counts=True)

    total = counts.sum(axis=0)
    counts = (counts/total)
    logs = np.log2(counts)
    counts = -1*counts

    return np.dot(counts,logs)


# classes : 
# 0 : Setosa
# 1 : Versicolor
# 2 : Virginica



def build_decision_tree(x_train,y_train,level,features):


    if len(set(y_train)) == 1 or (len(x_train[0])==1): # base case : either all values correspont to a single class or no features left to split on

        print('Level',level)
        print('This is a leaf node')
        v,c = np.unique(y_train,return_counts=True)
        
        for label,freq in zip(v,c):
            print('Count of',label,'=',freq)


        return

    ent = entropy(y_train)       # entropy before split

    info_gain = float('-inf')
    index = None   
    for i in range(len(x_train[0])):      # loop over all features, pick one which gives maximum information gain

        values,counts = np.unique(x_train[:,i],return_counts=True)

        total = counts.sum(axis=0)
        counts = counts/total          # weights for each of the split on a feature

        split_entropies = list()       

        for unique_val in values:

            v,c = np.unique(y_train[x_train[:,i]==unique_val],return_counts=True)
            tot = c.sum(axis=0)
            c = c/tot
            logs = np.log2(c)
            c = -1*c
            split_entropies.append(np.dot(c,logs))

        split_entropies = np.array(split_entropies)
        weighted_entropy = np.dot(counts,split_entropies)
        gain = ent - weighted_entropy

        if gain > info_gain:
            info_gain = gain
            index = i
    

    print('Level',level)
    y_labels,y_counts = np.unique(y_train,return_counts=True)
    for label,freq in zip(y_labels,y_counts):
        print('Count of',label,'=',freq)
    
    print('Current Entropy is:',ent)

    print('splitting on feature',features[index],'with info gain of',info_gain)
    temp = np.delete(features,index,0)              # remove the feature_name after split


    values = np.unique(x_train[:,index])
    new_train = np.delete(x_train,index,1)          # remove feature from dataset after split

    
    for split_point in values:

        bool = (x_train[:,index]==split_point)

        build_decision_tree(new_train[bool],y_train[bool],level+1,temp)
```

```python
build_decision_tree(arr,y_train,0,np.array(['sep_len','sep_wid','pet_len','pet_wid']))
```
